```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Cleaned_data.csv')

```

```python
df
```

```python
# Dropping column_area, story_number (Highly correlated)

df.drop(['storey_number'], axis=1, inplace=True)
df.drop(['Column_area'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y1 = le.fit_transform(y)
```

```python
y1
```

```python
# Transform to the one-hot encoding

y = pd.get_dummies(y1, dtype=int)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 56)
#X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# Feature scaling shoud be done after data splitting.There should be no influence on test data.

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train.iloc[:,:-1] = sc.fit_transform(X_train.iloc[:,:-1])

#Not using fit method so that model doesn't know the mean and standard deviation of the test dataset
X_test.iloc[:,:-1] = sc.transform(X_test.iloc[:,:-1])
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(14,)),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(5, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size=30, validation_split=0.2, epochs=30)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
```

```python
# Hyperparameter Tuning

# Keras Tuner 
```

```python
import keras_tuner as kt
```

```python
def build_model(hp):
    model = Sequential()
    model.add(keras.layers.Flatten(input_shape=(14,)))
              
    hp_activation = hp.Choice('activation', values=['relu','tanh'])
    hp_layer_1 = hp.Int('layer_1', min_value=1, max_value=20, step=2)
              
    model.add(Dense(units=hp_layer_1, activation=hp_activation))
    model.add(Dense(5, activation='softmax'))
    
    optimizer = hp.Choice('optimizer', values=['adam','sgd','rmsprop','adadelta'])
    
    model.compile(optimizer=optimizer , loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
tuner = kt.Hyperband(build_model,
                     objective='val_accuracy',
                     max_epochs=5,
                     factor=3,
                     directory='dir',
                     project_name='hyperparameter_tuning')
```

```python
tuner.search(X_train, y_train, epochs=5, validation_data=(X_test,y_test))
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
model = tuner.hypermodel.build(best_hps)
```

```python
history = model.fit(X_train, y_train, batch_size=30, validation_split=0.2, epochs=30)
```

```python
pd.DataFrame(history.history)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
```
