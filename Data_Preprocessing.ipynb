# Load_Data

# Data Preprocessing or Cleaning

# Dealing with missing data

```python
# Replacing non-standard value with 'Nan' so that pandas dataframe can ditect

df.replace('NA', np.nan, inplace = True)
df.replace('#VALUE!', np.nan, inplace = True)
df.replace(1999, np.nan, inplace = True)
df.replace('N/L', np.nan, inplace = True)
df.replace(0, np.nan, inplace = True)
```

```python
df.describe(include='all')
```

```python
# Dealing with those variables which need to be replaced by mean

from sklearn.impute import SimpleImputer
imputer_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
df[['Building_age','Overhanging_ratio']] = imputer_mean.fit_transform(df[['Building_age','Overhanging_ratio']]).astype('object')

```

```python
# Drop whole row with NaN in 'Class' column 
df.dropna(subset = ['Class'], axis=0, inplace=True)

# Reset index because of dropping rows
df.reset_index(drop = True, inplace = True)
```

# Outlier Detection

# Z score method

```python
# For normally distributed feature

print("Mean value of feature", df['Minimum_normalized_lateral_stiffness_index'].mean())
print("Std value of feature", df['Minimum_normalized_lateral_stiffness_index'].std())
print("Min value of feature", df['Minimum_normalized_lateral_stiffness_index'].min())
print("Max value of feature", df['Minimum_normalized_lateral_stiffness_index'].max())
```

```python
# Find the boundaries values
# Based on Z score calculation

print("Highest allowed", df['Minimum_normalized_lateral_stiffness_index'].mean() + 3*df['Minimum_normalized_lateral_stiffness_index'].std())
print("Lowest allowed", df['Minimum_normalized_lateral_stiffness_index'].mean() - 3*df['Minimum_normalized_lateral_stiffness_index'].std())
```

```python
# Find the outliers

df[(df['Minimum_normalized_lateral_stiffness_index']>1.1580185148669695) | (df['Minimum_normalized_lateral_stiffness_index'] < -0.8489075462582739)]
```

# IQR method (Interquartile Range)

```python
# For skewed feature (Priority_index)

df['Priority_index'].describe()
```

```python
sns.boxplot(df['Priority_index'])
```

```python
# Finding the IQR

percentile25 = df['Priority_index'].quantile(0.25)
percentile75 = df['Priority_index'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
iqr = percentile75 - percentile25 
```

```python
iqr
```

```python
# Indicating the limit values

upper_limit = percentile75 + 1.5*iqr
lower_limit = percentile25 - 1.5*iqr
```

```python
print('Upper limit', upper_limit)
print('Lower limit', lower_limit)
```

```python
# Find the outliers

df[df['Priority_index'] > upper_limit]
```

```python
df[df['Priority_index'] < lower_limit]
```

# Outlier Removal for Normally Distributed Feature

# Capping

```python
upper_limit = df['Minimum_normalized_lateral_stiffness_index'].mean() + 3*df['Minimum_normalized_lateral_stiffness_index'].std()
lower_limit = df['Minimum_normalized_lateral_stiffness_index'].mean() - 3*df['Minimum_normalized_lateral_stiffness_index'].std()
```

```python
df['Minimum_normalized_lateral_stiffness_index'] = np.where(
    df['Minimum_normalized_lateral_stiffness_index']> upper_limit, 
    upper_limit, 
    np.where(
        df['Minimum_normalized_lateral_stiffness_index'] < lower_limit,
        lower_limit, 
        df['Minimum_normalized_lateral_stiffness_index']
    )
)
```

```python
df['Minimum_normalized_lateral_stiffness_index'].describe()
```

# Outlier Removal for Skewed Feature

#  Capping

```python
# Priority_index

df['Priority_index'] = np.where(
    df['Priority_index']> upper_limit, 
    upper_limit, 
    np.where(
        df['Priority_index'] < lower_limit,
        lower_limit, 
        df['Priority_index']
    )
)
```
