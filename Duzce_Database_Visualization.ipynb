```python
cols_to_plot = ['Number_of_stories', 'Floor_area', 'Storey_height', 'Class']
```

```python
sns.pairplot(df[cols_to_plot])
```

```python
# Adding a regresssiopn line

sns.pairplot(df[cols_to_plot], diag_kind ='kde', kind='reg')
```

```python
#using plot_kws to change regression line color
sns.pairplot(df[cols_to_plot], diag_kind ='kde', kind='reg', plot_kws={'line_kws':{'color':'red'}})
```

```python
# Splitting data by category

sns.pairplot(df[cols_to_plot], hue='Class')
```

```python
g = sns.pairplot(df[cols_to_plot], kind='scatter', diag_kind='kde')

# remove upper triangle plots
for i, j in zip(*np.triu_indices_from(g.axes, 1)):
    g.axes[i, j].set_visible(False)
plt.show()
```

```python
from scipy.stats import pearsonr
def reg_coef(x,y,label=None,color=None,**kwargs):
    ax = plt.gca()
    r,p = pearsonr(x,y)
    ax.annotate('r = {:.2f}'.format(r), xy=(0.5,0.5), xycoords='axes fraction', ha='center')
    ax.set_axis_off()

g = sns.PairGrid(df[cols_to_plot])
g.map_diag(sns.distplot)
g.map_lower(sns.regplot)
g.map_upper(reg_coef)
```
